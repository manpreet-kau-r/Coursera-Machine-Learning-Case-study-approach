# using collaborative filtering and knn

```python
data=pd.read_csv('song_data.csv')
data.head()
```

```python
# most popular songs
data['song'].value_counts()
```

```python
df.loc['Kings Of Leon']
```

```python
# matrix of users*songs
song=data.drop_duplicates(['user_id','song','title','artist'])
```

```python
song.head()
```

```python
song.shape
```

```python
song=song.loc[0:50000,]
```

```python
song.shape
```

```python
matrix=song.pivot(index='user_id',columns='song_id',values='listen_count')
```

```python
matrix.head()
```

```python
matrix.shape
```

```python
matrix=matrix.fillna(0)
```

```python
matrix.head()
```

## nearest neighbor model

```python
# knn model using matrix
from sklearn.neighbors import NearestNeighbors
```

```python
knn=NearestNeighbors(n_neighbors=10,metric='cosine')
knn_model=knn.fit(matrix)
```

```python
# predict neighbors for a particular user
user=matrix.iloc[1,]
```

```python
distances,indices=knn_model.kneighbors([user])
print(distances)
print(indices)
```

```python
neighbors=[]
for i in indices[0][1:]:
    neighbors.append(matrix.iloc[i])
neighbors1=pd.DataFrame(neighbors)
neighbors1=np.array(neighbors1.index)
neighbors1=list(neighbors1)
neighbors1
```

```python
# similar users
neighbors=[]
for i in indices[0][1:]:
    neighbors.append(song[song.index==i].user_id.values[0])
neighbors
```

```python
neighbor_song=pd.DataFrame(columns=['user_id','song_id','listen_count','title','artist','song'])
for i in neighbors:
    neighbor_song=neighbor_song.append(song[song.user_id==i],ignore_index=True)
neighbor_song
```

```python
neighbor_song1=pd.DataFrame(columns=['user_id','song_id','listen_count','title','artist','song'])
for i in neighbors1:
    neighbor_song1=neighbor_song1.append(song[song.user_id==i],ignore_index=True)
neighbor_song1
```

```python
neighbor_songs=pd.DataFrame({'Count':neighbor_song['listen_count'],'Song':neighbor_song['song']})
neighbor_songs.head()
```

```python
# top songs listened by similar users
neighbor_songs=neighbor_songs.sort_values('Count',ascending=False)
neighbor_songs.head()
```

```python
neighbor_songs=neighbor_songs.reset_index(drop=True)
neighbor_songs.head()
```

```python
# recommend top 10 songs
Recommended_Song=neighbor_songs['Song'][0:10]
print(Recommended_Song)
```
