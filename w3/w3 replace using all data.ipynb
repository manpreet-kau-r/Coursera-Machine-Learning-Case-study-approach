```python
data=pd.read_csv('amazon_baby.csv')
data.head()
```

```python
#prod=data[data['name']=='Vulli Sophie the Giraffe Teether']
prod=data.copy()
prod.head()
```

## count words directly for selected words

```python
sel_words = ['awesome', 'great', 'fantastic', 'amazing', 'love', 'horrible', 'bad', 'terrible', 'awful', 'wow', 'hate']
cv=CountVectorizer(vocabulary=sel_words)
```

```python
prod['review'].fillna('0',inplace=True)
```

```python
cv_fit=cv.fit_transform(prod['review'])
words=cv.get_feature_names()
counts=cv_fit.toarray()
#cv.vocabulary_
```

```python
counts.shape
```

```python
#print(words)
len(prod)
```

```python
counts
```

```python
d=[]
for i in range(len(counts)):
    mapped=zip(words,counts[i])
    mapped=dict(mapped)
    d.append(mapped)
len(d)
```

```python
prod['spec_count']=d
prod.head()
```

```python
def awesome(d):
    if 'awesome' in d:
        return d['awesome']
    else:
        return 0
```

## seperate word count into columns

```python
def cols(d,word):
    if word in d:
        return d[word]
    else:
        return 0
```

```python
for word in sel_words:
    prod[word]=prod['spec_count'].apply(lambda d: cols(d,word))
prod.columns
```

```python
for word in sel_words:
    prod[word]=prod['spec_count'].apply(lambda d: d.get(word,0))
prod.columns
```

```python
prod.head()
```
